```python
import pandas as pd
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def ecludian(row):
  pt_A = row['1st point']
  print(pt_A)
  pt_B = row['2nd point']
  print(pt_B)

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
  return True
```

```python
points_df['distance']  = points_df.apply(ecludian,axis = 1)
points_df
```
